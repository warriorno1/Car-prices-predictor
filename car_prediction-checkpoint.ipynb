```python
import pandas as pd
```

```python
from sklearn import linear_model
```

```python
df = pd.read_csv('Downloads/carprices.csv')
```

```python
df
```

```python
df.columns = ['carModel','Mileage','sellPrice','Age']
```

```python
df
```

```python
dummy = pd.get_dummies(df.carModel)
```

```python
dummy
```

```python
merged = pd.concat([df,dummy],axis = 'columns')
```

```python
merged
```

```python
merged_new = merged.drop(['Mercedez Benz C class','carModel'],axis = 'columns')
```

```python
merged_new
```

```python
x = merged_new.drop(['sellPrice'],axis = 'columns')
```

```python
y = merged_new.sellPrice
```

```python
x
```

```python
y
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(x,y)
```

```python
reg.predict([[4500,4,0,0]])
```

```python
model = reg
```

```python
model.score(x,y)
```

```python
import pickle
```

```python
with open('model_pickle','wb') as f:
    pickle.dump(model,f)
```

```python
with open('model_pickle','rb') as f:
    mp = pickle.load(f)
```

```python
mp.predict([[3500,3,0,1]])
```
